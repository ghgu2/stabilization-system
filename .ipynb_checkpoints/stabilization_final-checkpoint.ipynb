```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
import os
import warnings
```

### Конфигурация

```python
warnings.filterwarnings('ignore')
INPUT_VIDEO = 'video.mp4'
OUTPUT_DIR  = 'result'
FB_PYR_SCALE  = 0.5
FB_LEVELS     = 3
FB_WIN_SIZE   = 15
FB_ITERATIONS = 3
FB_POLY_N     = 5
FB_POLY_SIGMA = 1.2
KALMAN_Q = 0.03   
KALMAN_R = 1500  
CROP_RATIO = 0.07  
os.makedirs(OUTPUT_DIR, exist_ok=True)
```

```python
warnings.filterwarnings('ignore')
```

### Импорт видео

```python
cap = cv2.VideoCapture(INPUT_VIDEO)

fps    = cap.get(cv2.CAP_PROP_FPS)
width  = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
total  = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
cap.release()


cap = cv2.VideoCapture(INPUT_VIDEO)
frames = []
while True:
    ret, frame = cap.read()
    if not ret:
        break
    frames.append(frame)
cap.release()

print(f'Разрешение: {width}×{height}')
print(f'FPS: {fps:.1f}')
print(f'Кадров: {total}')
print(f'Длит.: {total/fps:.1f}')
```

### Плотный оптический поток Farnebäck

```python
transforms = []
prev_gray = cv2.cvtColor(frames[0], cv2.COLOR_BGR2GRAY)

for i in tqdm(range(1, len(frames)), desc='Farnebäck'):
    curr_gray = cv2.cvtColor(frames[i], cv2.COLOR_BGR2GRAY)

    flow = cv2.calcOpticalFlowFarneback(
        prev_gray, curr_gray, None,
        FB_PYR_SCALE, FB_LEVELS, FB_WIN_SIZE,
        FB_ITERATIONS, FB_POLY_N, FB_POLY_SIGMA, 0
    )

    dx = float(np.median(flow[..., 0]))
    dy = float(np.median(flow[..., 1]))

    h_f, w_f = flow.shape[:2]
    left_dy  = float(np.median(flow[:, :w_f//2, 1]))
    right_dy = float(np.median(flow[:, w_f//2:, 1]))
    da = (right_dy - left_dy) / (w_f / 2 + 1e-6)

    transforms.append([dx, dy, da])
    prev_gray = curr_gray

transforms = np.array(transforms)
print(f'Трансформаций: {len(transforms)}')
print(f'Средние |dx|={np.mean(np.abs(transforms[:,0])):.2f}px')
print(f'|dy|={np.mean(np.abs(transforms[:,1])):.2f}px')
print(f'|da|={np.mean(np.abs(transforms[:,2])):.5f}рад')
```

### Сглаживание фильтром Калмана

```python
def kalman_smooth_1d(observed, Q=0.03, R=10.0):
    n = len(observed)
    F = np.array([[1, 1],
                  [0, 1]], dtype=np.float64)
    H = np.array([[1, 0]], dtype=np.float64)
    Q_mat = Q * np.eye(2)
    R_mat = np.array([[R]])

    x = np.array([observed[0], 0.0])
    P = np.eye(2) * 1.0

    smoothed = np.zeros(n)

    for i in range(n):
        x_pred = F @ x
        P_pred = F @ P @ F.T + Q_mat

        z = observed[i]
        y_res = z - (H @ x_pred)[0]
        S = (H @ P_pred @ H.T + R_mat)[0, 0]
        K = (P_pred @ H.T) / S

        x = x_pred + K.flatten() * y_res
        P = (np.eye(2) - K @ H) @ P_pred

        smoothed[i] = x[0]

    return smoothed


trajectory = np.cumsum(transforms, axis=0)

smoothed = np.zeros_like(trajectory)
for col in range(3):
    smoothed[:, col] = kalman_smooth_1d(
        trajectory[:, col], Q=KALMAN_Q, R=KALMAN_R
    )

correction = smoothed - trajectory

print(f'Калман-сглаживание (Q={KALMAN_Q}, R={KALMAN_R})')
print(f'Макс. коррекция: dx={np.max(np.abs(correction[:,0])):.1f}px')
print(f'dy={np.max(np.abs(correction[:,1])):.1f}px')
```

### Warp кадров

```python
stabilized = [frames[0].copy()]

for i in tqdm(range(len(correction)), desc='Warp кадров'):
    dx, dy, da = correction[i]

    M = np.float32([
        [ np.cos(da), -np.sin(da), dx],
        [ np.sin(da),  np.cos(da), dy]
    ])

    warped = cv2.warpAffine(
        frames[i + 1], M, (width, height),
        flags=cv2.INTER_LINEAR,
        borderMode=cv2.BORDER_REPLICATE
    )
    stabilized.append(warped)
```

### Примеры warping-а

```python
warp_indices = np.linspace(0, len(correction) - 1, 5, dtype=int).tolist()

n = len(warp_indices)
fig, axes = plt.subplots(n, 2, figsize=(14, 4.5 * n))
axes[0, 0].set_title('Оригинальный кадр', fontsize=12, fontweight='bold')
axes[0, 1].set_title('После warpAffine (до crop)', fontsize=12, fontweight='bold')

for row, ci in enumerate(warp_indices):
    frame_idx = ci + 1 
    dx, dy, da = correction[ci]

    M = np.float32([
        [ np.cos(da), -np.sin(da), dx],
        [ np.sin(da),  np.cos(da), dy]
    ])
    warped = cv2.warpAffine(
        frames[frame_idx], M, (width, height),
        flags=cv2.INTER_LINEAR,
        borderMode=cv2.BORDER_REPLICATE
    )


    axes[row, 0].imshow(cv2.cvtColor(frames[frame_idx], cv2.COLOR_BGR2RGB))
    axes[row, 0].set_ylabel(
        f'Кадр {frame_idx}\n'
        f'dx={dx:+.1f} px\n'
        f'dy={dy:+.1f} px\n'
        f'da={da:+.4f} рад',
        fontsize=9
    )
    axes[row, 0].axis('off')


    axes[row, 1].imshow(cv2.cvtColor(warped, cv2.COLOR_BGR2RGB))
    axes[row, 1].axis('off')

plt.tight_layout()
plt.show()
```

### Сrop краевых артефактов и resize

```python
cx = int(width  * CROP_RATIO)
cy = int(height * CROP_RATIO)
crop_w = width  - 2 * cx
crop_h = height - 2 * cy


cropped = []
for frame in tqdm(stabilized, desc='Crop + resize'):
    roi = frame[cy:height-cy, cx:width-cx]
    resized = cv2.resize(roi, (width, height), interpolation=cv2.INTER_LINEAR)
    cropped.append(resized)
```

### Сохранение обработанного видео

```python
out_path = os.path.join(OUTPUT_DIR, 'stabilized_farneback.mp4')
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter(out_path, fourcc, fps, (width, height))
for f in cropped:
    out.write(f)
out.release()

size_mb = os.path.getsize(out_path) / (1024 * 1024)
```

### Сравнение траекторий движения камеры

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 4))
fig.suptitle(f'Траектория камеры: оригинал vs Калман (Q={KALMAN_Q}, R={KALMAN_R})',
             fontsize=13, fontweight='bold')

labels = ['dx (горизонталь, px)', 'dy (вертикаль, px)', 'da (поворот, рад)']

for col, label in enumerate(labels):
    ax = axes[col]
    ax.plot(trajectory[:, col], color='#e74c3c', alpha=0.7, linewidth=1, label='Оригинал')
    ax.plot(smoothed[:, col],   color='#2980b9', alpha=0.9, linewidth=1.5, label='Калман')
    ax.set_title(label)
    ax.set_xlabel('Кадр')
    ax.legend(fontsize=8)
    ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, 'trajectories.png'), dpi=150, bbox_inches='tight')
plt.show()
```

### Сравнение кадров оригинала и обработанного изображения

```python
sample_indices = np.linspace(0, len(frames) - 1, 5, dtype=int).tolist()

n = len(sample_indices)
fig, axes = plt.subplots(n, 2, figsize=(12, 4 * n))
axes[0, 0].set_title('Оригинал', fontsize=12, fontweight='bold')
axes[0, 1].set_title('Стабилизированный (Farnebäck + Калман + crop)', fontsize=12, fontweight='bold')

for row, idx in enumerate(sample_indices):
    axes[row, 0].imshow(cv2.cvtColor(frames[idx], cv2.COLOR_BGR2RGB))
    axes[row, 0].set_ylabel(f'Кадр {idx}', fontsize=9)
    axes[row, 0].axis('off')

    axes[row, 1].imshow(cv2.cvtColor(cropped[idx], cv2.COLOR_BGR2RGB))
    axes[row, 1].axis('off')

plt.tight_layout()
plt.show()
```

###  Поле оптического потока Farnebäck

```python
mid = len(frames) // 2
gray1 = cv2.cvtColor(frames[mid], cv2.COLOR_BGR2GRAY)
gray2 = cv2.cvtColor(frames[mid + 1], cv2.COLOR_BGR2GRAY)

flow = cv2.calcOpticalFlowFarneback(
    gray1, gray2, None,
    FB_PYR_SCALE, FB_LEVELS, FB_WIN_SIZE,
    FB_ITERATIONS, FB_POLY_N, FB_POLY_SIGMA, 0
)

step = 16
h_f, w_f = gray1.shape
y, x = np.mgrid[step//2:h_f:step, step//2:w_f:step]
fx = flow[y, x, 0]
fy = flow[y, x, 1]

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Стрелки поверх кадра
axes[0].imshow(cv2.cvtColor(frames[mid], cv2.COLOR_BGR2RGB))
axes[0].quiver(x, y, fx, fy, color='lime', scale=50, width=0.002, alpha=0.8)
axes[0].set_title(f'Поле потока (кадр {mid} → {mid+1})', fontsize=12)
axes[0].axis('off')

# HSV-визуализация
magnitude, angle = cv2.cartToPolar(flow[..., 0], flow[..., 1])
hsv = np.zeros((*gray1.shape, 3), dtype=np.uint8)
hsv[..., 0] = angle * 180 / np.pi / 2
hsv[..., 1] = 255
hsv[..., 2] = cv2.normalize(magnitude, None, 0, 255, cv2.NORM_MINMAX)
flow_rgb = cv2.cvtColor(hsv, cv2.COLOR_HSV2RGB)
axes[1].imshow(flow_rgb)
axes[1].set_title('HSV-карта (цвет=направление, яркость=скорость)', fontsize=12)
axes[1].axis('off')

plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, 'flow_field.png'), dpi=150, bbox_inches='tight')
plt.show()
```

### Метрики стабилизации

```python
orig_jitter = np.mean(np.abs(transforms), axis=0)  

stab_transforms = []
prev_gray_s = cv2.cvtColor(cropped[0], cv2.COLOR_BGR2GRAY)
for i in tqdm(range(1, len(cropped)), desc='Поток (стаб.)'):
    curr_gray_s = cv2.cvtColor(cropped[i], cv2.COLOR_BGR2GRAY)
    flow_s = cv2.calcOpticalFlowFarneback(
        prev_gray_s, curr_gray_s, None,
        FB_PYR_SCALE, FB_LEVELS, FB_WIN_SIZE,
        FB_ITERATIONS, FB_POLY_N, FB_POLY_SIGMA, 0
    )
    sdx = float(np.median(flow_s[..., 0]))
    sdy = float(np.median(flow_s[..., 1]))
    h_s, w_s = flow_s.shape[:2]
    sda = (float(np.median(flow_s[:, w_s//2:, 1])) - float(np.median(flow_s[:, :w_s//2, 1]))) / (w_s/2 + 1e-6)
    stab_transforms.append([sdx, sdy, sda])
    prev_gray_s = curr_gray_s

stab_transforms = np.array(stab_transforms)
stab_jitter = np.mean(np.abs(stab_transforms), axis=0)

def interframe_diff(frame_list):
    diffs = []
    for i in range(1, len(frame_list)):
        d = cv2.absdiff(frame_list[i], frame_list[i - 1])
        diffs.append(float(np.mean(d)))
    return np.array(diffs)

orig_diff = interframe_diff(frames)
stab_diff = interframe_diff(cropped)


print()
print('=' * 55)
print(f"{'Метрика':<35} {'Ориг':>8} {'Стаб':>8}")
print('-' * 55)
print(f"{'Среднее |dx| (px)':<35} {orig_jitter[0]:>8.2f} {stab_jitter[0]:>8.2f}")
print(f"{'Среднее |dy| (px)':<35} {orig_jitter[1]:>8.2f} {stab_jitter[1]:>8.2f}")
print(f"{'Среднее |da| (рад)':<35} {orig_jitter[2]:>8.5f} {stab_jitter[2]:>8.5f}")
print(f"{'Межкадр. различие (mean px)':<35} {np.mean(orig_diff):>8.2f} {np.mean(stab_diff):>8.2f}")
print(f"{'Межкадр. различие (std px)':<35} {np.std(orig_diff):>8.2f} {np.std(stab_diff):>8.2f}")
print('=' * 55)


reduction_dx = (1 - stab_jitter[0] / (orig_jitter[0] + 1e-9)) * 100
reduction_dy = (1 - stab_jitter[1] / (orig_jitter[1] + 1e-9)) * 100
print(f'\nСнижение дрожания: dx на {reduction_dx:.1f}%, dy на {reduction_dy:.1f}%')
```

### Межкадровое различие

```python
fig, axes = plt.subplots(2, 1, figsize=(14, 7))

# Межкадровое различие
axes[0].plot(orig_diff, color='#e74c3c', alpha=0.7, linewidth=1, label='Оригинал')
axes[0].plot(stab_diff, color='#3498db', alpha=0.8, linewidth=1, label='Стабилизированный')
axes[0].set_title('Межкадровое различие (меньше = плавнее)', fontsize=12)
axes[0].set_xlabel('Кадр')
axes[0].set_ylabel('Средняя разница (px)')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

# Амплитуда межкадровых смещений
orig_amp = np.sqrt(transforms[:, 0]**2 + transforms[:, 1]**2)
stab_amp = np.sqrt(stab_transforms[:, 0]**2 + stab_transforms[:, 1]**2)
axes[1].plot(orig_amp, color='#e74c3c', alpha=0.7, linewidth=1, label='Оригинал')
axes[1].plot(stab_amp, color='#3498db', alpha=0.8, linewidth=1, label='Стабилизированный')
axes[1].set_title('Амплитуда межкадрового смещения (px)', fontsize=12)
axes[1].set_xlabel('Кадр')
axes[1].set_ylabel('Амплитуда (px)')
axes[1].legend()
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, 'metrics_comparison.png'), dpi=150, bbox_inches='tight')
plt.show()
```
